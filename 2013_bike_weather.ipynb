```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
from zipfile import ZipFile
```

```python
source = '/Users/jonleckie/Desktop/DSI_all/capstones/capstone_one/'
```

```python
for item in os.listdir(source):
    if item.endswith('.zip'):
        file_name = source + item
        zip_ref = ZipFile(file_name)
        zip_ref.extractall(source)
        zip_ref.close()
```

```python
os.chdir(source)
```

```python
extension = 'csv'
all_files = [i for i in glob.glob('*.{}'.format(extension))]

bike_files = []
for file in all_files:
    if file.startswith('2'):
        bike_files.append(file)
        
bike_files
```

```python
bike_file = pd.concat([pd.read_csv(f) for f in bike_files])
bike_file.to_csv('bike_trips_13.csv', index=False, encoding='utf-8-sig')
```
